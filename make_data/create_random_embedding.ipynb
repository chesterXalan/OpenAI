```python
import json
import random
```

## training data

```python
random.seed(0)
all_data = []

for i in range(10): # num of class
    rand_range1, rand_range2 = i/10, (i+1)/10
    for _ in range(100): # num of data
        data = ''
        for _ in range(128): # len of embedding
            data += f'{random.uniform(rand_range1, rand_range2):.4f} '
            
        data_dict = {'prompt': data.rstrip(), 'completion': f'{i}'}
        all_data.append(data_dict)
random.shuffle(all_data)
```

```python
all_data[:10]
```

```python
with open(r'..\data\random_embedding-training.jsonl', 'w') as f:
    for dict_ in all_data:
        json.dump(dict_, f)
        f.write('\n')
```

## testing data

```python
random.seed(1)
all_data = []

for i in range(10): # num of class
    rand_range1, rand_range2 = i/10, (i+1)/10
    for _ in range(50): # num of data
        data = ''
        for _ in range(128): # len of embedding
            data += f'{random.uniform(rand_range1, rand_range2):.4f} '
            
        data_dict = {'prompt': data.rstrip(), 'completion': f'{i}'}
        all_data.append(data_dict)
```

```python
with open(r'..\data\random_embedding-testing.jsonl', 'w') as f:
    for dict_ in all_data:
        json.dump(dict_, f)
        f.write('\n')
```
